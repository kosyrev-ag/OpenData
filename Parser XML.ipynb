```python
import xml.etree.ElementTree as Xet
import pandas as pd
import os
```

```python
directory = "data/xml/"
files = os.listdir(directory)
```

```python
files
```

```python
cols = ['ID Документа', 'Дата докумекента', 'Дата составленгия документа', 'Наименование организации', 'ИНН', 'Количество работников']
rows = []
```

```python
for file in files:
    p_file = directory + '/' + file
    xmlparse = Xet.parse(p_file)
    root = xmlparse.getroot()
    i = 0
    for child in root:
        if i == 0:
            i = i+1
            continue
        id_doc = child.attrib['ИдДок']
        date_doc = child.attrib['ДатаДок']
        date_wr = child.attrib['ДатаСост']
        name = child[0].attrib['НаимОрг']
        inn = child[0].attrib['ИННЮЛ']
        count = int(child[1].attrib['КолРаб'])
        rows.append({"ID Документа": id_doc,
                     "Дата докумекента": date_doc,
                     "Дата составленгия документа": date_wr,
                     "Наименование организации": name,
                     "ИНН": inn,
                     "Количество работников": count})
```

```python
df = pd.DataFrame(rows, columns=cols)         
```

```python
df
```

```python
df.to_csv('data/csv/workers.csv', sep=';', encoding='cp1251')
```
